```python
import requests
import pandas as pd
import json
import time
import warnings
import re
warnings.simplefilter(action='ignore', category=FutureWarning)

# Where we shall store the data
df = pd.DataFrame()

IMAGE_PATH = "Images/"


def json_to_df(test_df, json):
    """
    Adds and converts MTG Json into Dataframe
    """
    for i in json:
        name = i["name"]
        print(name)
        
        if "//" in name:
            for j in i["card_faces"]:

                name = j["name"]
                try:
                    colors = j["colors"]
                except:
                    colors = i["colors"]

                cmc = i["cmc"]

                try:
                    power_and_toughness = int(j["power"]) + int(j["toughness"])
                except:
                    power_and_toughness = 0

                release_date = i["released_at"]
                keywords = i["keywords"]

                # Saves images
                time.sleep(0.1)
                try:
                    image_uri = j["image_uris"]["normal"]
                except:
                    image_uri = i["image_uris"]["normal"]
                save_image(IMAGE_PATH, image_uri, name)
                
                added_row = {"name": name,
                         "colors": colors, 
                         "cmc": cmc, 
                         "power_and_toughness": power_and_toughness, 
                         "release_date": release_date, 
                         "keywords": keywords}

                test_df = test_df.append(added_row, ignore_index = True)
        
        else:
            name = i["name"]
            colors = i["colors"]

            cmc = i["cmc"]

            try:
                power_and_toughness = int(i["power"]) + int(i["toughness"])
            except:
                power_and_toughness = 0

            release_date = i["released_at"]
            keywords = i["keywords"]

            # Saves images
            time.sleep(0.1)
            image_uri = i["image_uris"]["normal"]
            save_image(IMAGE_PATH, image_uri, name)

            added_row = {"name": name,
                         "colors": colors, 
                         "cmc": cmc, 
                         "power_and_toughness": power_and_toughness, 
                         "release_date": release_date, 
                         "keywords": keywords}

            test_df = test_df.append(added_row, ignore_index = True)
        
    # You MUST return the dataframe at the end
    return test_df

def save_image(path, url, name):
    response = requests.get(url)
    
    # Removes all non-alphanumeric characters
    with open('{}{}.png'.format(path, re.sub(r'\W+', '_', name)), 'wb') as f:
        f.write(response.content)
```

```python
# Request data with the appropriate limitations
response = requests.get("https://api.scryfall.com/cards/search?q=is%3Acommander+not%3Afunny+in%3Apaper&unique=cards")

while True:
    time.sleep(0.2)
    data = response.json()["data"]
    df = json_to_df(df, data)
    
    if not response.json()["has_more"]:
        break
        
    response = requests.get(response.json()["next_page"])
    
    
```

```python
df.to_csv("commander_data.csv")
```
