```python
import requests
import pandas as pd
import json
import time

# Where we shall store the data
df = pd.DataFrame()

def json_to_df(test_df, json):
    """
    Adds and converts MTG Json into Dataframe
    """
    for i in json:
        name = i["name"]
        try:
            colors = i["colors"]
        except:
            colors = ["N"]
        cmc = i["cmc"]
        try:
            power_and_toughness = int(i["power"]) + int(i["toughness"])
        except:
            power_and_toughness = 0
        release_date = i["released_at"]
        keywords = i["keywords"]

        added_row = {"name": name,
                     "colors": colors, 
                     "cmc": cmc, 
                     "power_and_toughness": power_and_toughness, 
                     "release_date": release_date, 
                     "keywords": keywords}
        
        test_df = test_df.append(added_row, ignore_index = True)
    
    # You MUST return the dataframe at the end
    return test_df

```

```python
# Request data with the appropriate limitations
response = requests.get("https://api.scryfall.com/cards/search?q=is%3Acommander+not%3Afunny+in%3Apaper&unique=cards")
count = 0
while True:
    time.sleep(0.2)
    count += 1
    print(count)
    data = response.json()["data"]
    df = json_to_df(df, data)
    
    if not response.json()["has_more"]:
        break
        
    response = requests.get(response.json()["next_page"])
    
    
```

```python
df.to_csv("commander_data.csv")
```
